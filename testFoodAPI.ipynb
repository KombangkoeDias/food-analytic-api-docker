```python
import numpy as np

image_file = 'imgs/test.jpg'

files = {"image": open(image_file, 'rb')}
response = requests.post(f"{endpoint}/inference", files=files)
try:
    data = response.json()
    print(np.unique(data['prediction']))                
except requests.exceptions.RequestException:
    print(response.text)
```

```python
from PIL import Image
from io import BytesIO
import base64
image = data['visualization']
Image.open(BytesIO(base64.b64decode(image)))
```
